```python
### Generate ROC analysis and visualization in Google Colab with Python
# Import df.bl data, generated in R, from my drive
from google.colab import drive
drive.mount('/content/drive')
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from pingouin import partial_corr

df_bl = pd.read_csv("drive/MyDrive/df_bl.csv")
```

```python
df_bl['AGE'] = pd.to_numeric(df_bl['AGE'], errors='coerce')
df_bl['PTEDUCAT'] = pd.to_numeric(df_bl['PTEDUCAT'], errors='coerce')
df_bl['APOE4'] = pd.to_numeric(df_bl['APOE4'], errors='coerce')
df_bl['PTGENDER'] = pd.factorize(df_bl['PTGENDER'])[0]
```

```python
# Calculate partial Spearman rank correlation between GAP.43 and PLASMA_NFL
cor_result_GAP_43_PLASMA_NFL = partial_corr(df_bl, x='GAP.43', y='PLASMA_NFL', covar=['AGE', 'PTEDUCAT', 'PTGENDER', 'APOE4'], method='spearman')

# Create scatter plot with partial correlation and grouped by DX
plt.figure(figsize=(6, 5))
sns.regplot(data=df_bl, x='GAP.43', y='PLASMA_NFL', color='blue', scatter_kws={'alpha': 0.5})
sns.lmplot(data=df_bl, x='GAP.43', y='PLASMA_NFL', hue='DX', palette=['#825C9F', '#9CCDBA', '#F1A84A'], ci=None)
plt.title('GAP.43 vs PLASMA_NFL')
plt.xlabel('GAP.43')
plt.ylabel('PLASMA_NFL')
plt.show()
plt.close()
print(cor_result_GAP_43_PLASMA_NFL)
```
